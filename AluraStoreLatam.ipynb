### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_t4 = tienda4['Categoría del Producto'].value_counts()

ventas_categoria_t1
```

**INFORME FINAL — Challenge Alura Store**

En este análisis utilicé los datos de ventas de las cuatro tiendas de Alura Store con el objetivo de identificar cuál de ellas sería la más conveniente para vender. Para ello, analicé la facturación total, las ventas por categoría, la calificación promedio de los clientes, los productos más y menos vendidos y el costo de envío promedio por tienda, utilizando Python y la librería pandas.

**Análisis de facturación**

Al analizar la facturación total de cada tienda, observé que la Tienda 1 presenta la mayor facturación, seguida por la Tienda 2 y la Tienda 3. Por otro lado, la Tienda 4 es la que registra la menor facturación total.

Esto indica que la Tienda 1 tiene un mayor volumen de ingresos, mientras que la Tienda 4 muestra un desempeño económico más bajo en comparación con las demás.

**Ventas por categoría**

El análisis de ventas por categoría muestra que las categorías Muebles y Electrónicos son las más vendidas, concentrando la mayor parte de las ventas en las tiendas. En cambio, categorías como Libros y Artículos para el hogar presentan un menor volumen de ventas.

Esto sugiere que la demanda de los clientes se orienta principalmente a productos de mayor valor y uso cotidiano, mientras que algunas categorías tienen menor rotación.

**Calificación promedio de las tiendas**

Al calcular la calificación promedio, se observa que todas las tiendas mantienen valores similares, lo que indica un nivel de satisfacción aceptable por parte de los clientes. Sin embargo, algunas tiendas presentan una calificación levemente inferior, lo que podría impactar en la percepción del servicio y en futuras ventas.

**Productos más y menos vendidos**

Cada tienda cuenta con productos que destacan por su alta cantidad de ventas, lo que demuestra una preferencia clara de los clientes por determinados artículos. Al mismo tiempo, también se identificaron productos con muy baja rotación, lo que podría indicar problemas de demanda, precio o falta de interés del público.

Este análisis permite detectar oportunidades para optimizar el catálogo de productos y mejorar la gestión del stock.

**Costo de envío promedio**

El costo de envío promedio varía entre las tiendas. En particular, la Tienda 4 presenta un costo de envío relativamente alto en relación con su facturación, lo que puede influir negativamente en la decisión de compra de los clientes y afectar el volumen de ventas.

**Conclusión final**

Teniendo en cuenta todos los análisis realizados, considero que la Tienda 4 es la opción más conveniente para vender, ya que presenta la menor facturación total y no se destaca significativamente en otros indicadores clave como ventas por categoría o satisfacción del cliente.

En cambio, la Tienda 1 muestra un mejor desempeño general, con mayor facturación y buen volumen de ventas, por lo que resulta más conveniente mantenerla en operación.

Este análisis demuestra cómo el uso de datos permite tomar decisiones estratégicas fundamentadas y reducir la incertidumbre al momento de evaluar el rendimiento de un negocio.
